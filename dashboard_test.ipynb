```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

def convert_fractional_date(date_str):
    if '.' not in date_str:
        raise ValueError("date format expected : 'YYYY-MM-DD.DD'")
    
    date_part, fraction_part = date_str.split('.')
    
    base_date = datetime.strptime(date_part, "%Y-%m-%d")
    
    fraction = float("0." + fraction_part)
    added_seconds = fraction * 24 * 60 * 60
    
    final_date = base_date + timedelta(seconds=added_seconds)
    return final_date

target_date = convert_fractional_date(df['date'][0])

current_date = datetime.now()

time_difference = relativedelta(target_date, current_date)

print(f"Time remaining: {time_difference.years} years, {time_difference.months} months, {time_difference.days} days, {time_difference.hours} hours, {time_difference.minutes} minutes, {time_difference.seconds} seconds")
```
