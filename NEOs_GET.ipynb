```python
import requests
import pandas as pd
```

```python
from astroquery.jplhorizons import Horizons

# ID de l'astéroïde et date d'observation
obj = Horizons(id='2025 FC2', location='500@10', epochs={'start': '2025-03-23', 'stop': '2027-05-23', 'step': '1d'})

# Récupérer les coordonnées héliocentriques
vectors = obj.vectors()

# Afficher les résultats
print(vectors['x', 'y', 'z'])  # Coordonnées en AU (Astronomical Units)

#convert en km
vectors['x'] = vectors['x']*149597870.7
vectors['y'] = vectors['y']*149597870.7
vectors['z'] = vectors['z']*149597870.7
```
