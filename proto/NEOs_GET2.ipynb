```python
import requests
import pandas as pd
```

```python
url = f'https://ssd-api.jpl.nasa.gov/sentry.api'
r = requests.get(url)
data = r.json()['data']
print(len(data))
```

```python
for item in data:
    print(f'name: {item['des']}')
    print(f'\tPalermo Scale: {item["ps_cum"]}')
    print(f'\tTorino Scale: {item["ts_max"]}')
    print()
```
